```python
au_train = df_train.replace(' ?',ny.nan)
```

```python
train.isnull().sum()
```

```python
plt.figure(figsize=(4,4))
sns.histplot(train['class'])
```

From the above we can conclude that in our training dataset has majority of people earn more than 50K (around 22000) and less than 50K (around 10000)


```python
sns.barplot(y=train['hours-per-week'],x=train['class'],data=train)
```

From this barplot between hours-per-week vs class:

   **we can understand that working more hours does not mean higher pay** 
       

```python
sns.jointplot(train['age'],train['hours-per-week'],kind='kde')
```

  This is the joint plot between age vs hours-per-week:
  
   **Majority of people for 40 hours a week irrespective of age group**
   
   
   **Maximum people either retire or doesnt work after the age of 70**
   
   
   **People in States start to work around the age of 10-15**
